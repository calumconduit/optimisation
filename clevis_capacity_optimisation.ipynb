```python
import numpy as np

from scipy.optimize import minimize as minimise
from scipy import optimize
```

```python
F_ty = 400
```

```python
K = 1.6
```

```python
def capacity(x, F_ty, K):
    t = x[0]
    D = x[1]
    w = x[2]
    bending_capacity = (np.pi * D ** 3 * F_ty)/ (16 * t)
    bearing_capacity = K * F_ty * D * t
    section_capacity = F_ty * (w-D) * t
    # print(f'({bending_capacity}, {bearing_capacity}, {section_capacity})')
    return -min(bending_capacity, bearing_capacity, section_capacity)
```

```python
rranges = (slice(0.1, 10, 1), slice(0.1, 15, 1), slice(0.1, 15, 1))
result = optimize.brute(capacity, rranges, args=(F_ty, K))
result[0]
```
